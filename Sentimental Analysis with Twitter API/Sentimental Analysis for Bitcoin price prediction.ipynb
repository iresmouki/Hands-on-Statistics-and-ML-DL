```python
## Load the text file and convert all text into lower case for easy processing of content

def dataprocessing(textfilename):
    tweets_data = []
    tweets_file = open(textfilename, "r")
    for line in tweets_file:
        try:
            tweet = json.loads(line.lower())
            tweets_data.append(tweet)
        except:
            continue
    pd.set_option('display.max_colwidth', -1)  ## To set maximum column size to the data frame c
    data = pd.DataFrame()  ## Creating a new data frame which contains the entire test data
    tweetslist = []
    tweetsdate = []
     
    ## Analyse the test data and create two new columns for the tweets and its respective location
    for i in range(len(tweets_data)):
        tweetslist.append(tweets_data[i]['full_text'])
        tweetsdate.append(tweets_data[i]['created_at'])
    
    data['Tweet_Text'] = tweetslist
    data['Date'] = tweetsdate
    data['Date'] = pd.to_datetime(data['Date']).dt.date
    
    return data
```

# 2. Sentiment Analysis : textBlob

```python
# Sentiment analysis using textblob :  clean tweets, classify sentiment, calculate polarity 
def clean_tweet(tweet):
    '''
    Utility function to clean the text in a tweet by removing 
    links and special characters using regex.
    '''
    tweet1 = ' '.join(re.sub("(@[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)", " ", tweet).split())
    return tweet1.lower()

def binary_analize_sentiment(tweet):
    '''
    Utility function to classify the polarity of a tweet
    using textblob.
    '''
    analysis = TextBlob(clean_tweet(str(tweet)))
    if analysis.sentiment.polarity > 0:
        return 'POS'
    elif analysis.sentiment.polarity == 0:
        return 'N'
    else:
        return 'NEG'
    
    
def polarity(tweet): 
    '''
    Utility function to calculate the polarity of a tweet
    using textblob.
    '''
    analysis = TextBlob(clean_tweet(str(tweet)))
    return analysis.sentiment.polarity
```

```python
## calls the above function to perform sentiment analysis
## Saves the result to a new column names 'SA' and 'polarity'

Data['polarity'] = np.array([ polarity(tweet) for tweet in Data['Tweet_Text'] ])
Data['SA'] = np.array([ binary_analize_sentiment(tweet) for tweet in Data['Tweet_Text'] ])

# We display the updated dataframe with the new column:
display(Data.head(10))
```

```python
df1 = Data.pivot_table(index='Date', values=['polarity'],columns=['SA'], aggfunc='mean')
df1.columns = ['pol_N','pol_NEG','pol_POS']
df1
```

```python
df2 = Data[['Date','SA']].pivot_table(index='Date', columns=['SA'], aggfunc=len)
df2.columns = ['N','NEG','POS']
for i in df2.columns:
    df2[i+'h'] = df2[i]/(df2['N']+df2['NEG']+df2['POS'])
df2 = df2.drop(columns=['N','NEG','POS'], axis=1)
df2.columns = ['ratio_N','ratio_NEG','ratio_POS']
df2
```

```python
df = pd.merge(df1,df2, how='left', on='Date')
df
```
